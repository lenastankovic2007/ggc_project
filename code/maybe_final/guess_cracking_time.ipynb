```python
import random

# Lozinke i njihova stvarna vremena krekovanja
password_data = {
    "Str@wBleki14!": "5 minutes",
    "Bleki_L0v3sT3nnis": "22 minutes",
    "StrawDog_14Serb!": "2 days",
    "strawberry2017": "less than a second",
    "strawberry17": "less than a second",
    "sara17": "less than a second"
}

print("üîê Password Cracking Match Game\n")
print("üéØ Uparite lozinke sa procenjenim vremenom krekovanja (offline, fast hashing).")
print("‚ÑπÔ∏è Svaka vremenska opcija se mo≈æe koristiti samo JEDNOM.\n")

# Izme≈°aj lozinke
passwords = list(password_data.keys())
random.shuffle(passwords)

# Kreiraj listu vremenskih vrednosti (ƒçak i ako su iste)
correct_times = [password_data[pwd] for pwd in passwords]
shuffled_times = correct_times.copy()
random.shuffle(shuffled_times)

# Obezbedi da svaka vremenska opcija bude jedinstvena ‚Äì ƒçak i ako je tekst isti
# Time se izbegava problem duplikata pri odabiru
numbered_options = {}
used_options = set()

for idx, t in enumerate(shuffled_times):
    key = str(idx + 1)
    # Dodaj oznaku ako se vrednost ponavlja
    count = list(shuffled_times[:idx]).count(t)
    display_value = f"{t}" if count == 0 else f"{t} ({count+1})"
    numbered_options[key] = display_value

# Prika≈æi sve lozinke
print("üìù Lozinke za povezivanje:")
for i, pwd in enumerate(passwords, 1):
    print(f"  {i}. {pwd}")
print()

# Prika≈æi sve vremenske opcije sa numeracijom
print("‚è± Vremenske opcije:")
for num, t in numbered_options.items():
    print(f"  {num}. {t}")
print("\nüß© Poƒçnimo!\n")

# Evidencija korisniƒçkih odgovora
user_answers = {}

for i, pwd in enumerate(passwords):
    while True:
        print(f"\nüîë Lozinka {i+1}/{len(passwords)}: {pwd}")
        print("‚¨áÔ∏è Preostale opcije:")
        for num, t in numbered_options.items():
            if num not in used_options:
                print(f"  {num}. {t}")
        choice = input("üëâ Izaberi broj vremena: ").strip()
        if choice not in numbered_options:
            print("‚ö†Ô∏è Neva≈æeƒái unos. Poku≈°aj ponovo.")
        elif choice in used_options:
            print("‚ö†Ô∏è Ta opcija je veƒá iskori≈°ƒáena.")
        else:
            used_options.add(choice)
            user_answers[pwd] = numbered_options[choice].split(" (")[0]  # ƒçista vrednost
            break

# Evaluacija
print("\nüìä Rezultati:\n")
score = 0
for pwd in passwords:
    correct = password_data[pwd]
    chosen = user_answers[pwd]
    match = "‚úÖ" if correct == chosen else "‚ùå"
    print(f"{match} Lozinka: {pwd}")
    print(f"   ‚û§ Tvoj odgovor: {chosen}")
    print(f"   ‚úî Taƒçno:        {correct}\n")
    if correct == chosen:
        score += 1

print(f"üéØ Taƒçnih odgovora: {score} od {len(passwords)}")
if score == len(passwords):
    print("üéâ Fantastiƒçno! Sve taƒçno!")
elif score >= 4:
    print("üëç Vrlo dobro!")
else:
    print("üìò Vredi probati jo≈° jednom.")


```
